```python
provinces.append(("Habana", 728.26, 2132183))
for n, d, p in provinces:
    t = d/proportionality
    print(n, '->', t, '->', t-p)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('dark_background')


pos = [(n, p - d/proportionality if d/proportionality - p > 0 else 0) for n, d, p in provinces]
neg = [(n, p - d/proportionality  if d/proportionality - p < 0 else 0) for n, d, p in provinces]
X = np.array([n for n, _ in pos])
Y1 = np.array([n for _, n in pos])
Y2 = np.array([n for _, n in neg])

fig, ax = plt.subplots()
ax.set_ylabel('Provincias', {"fontsize": 15})
ax.set_title('Diferencia entre población real e ideal', {"fontsize": 15}, pad=20, loc='left')
ax.barh(X, Y1, facecolor='red', edgecolor='white')
ax.barh(X, Y2, facecolor='green', edgecolor='white')


ax.grid(True, linestyle='-')
plt.show()
```
